The final modeling phase will test and compare various predictive algorithms. Moving forward, more complex non-linear and tree-based models such as Random Forest Regressor and XGBoost will be implemented and optimized via extensive hyperparameter tuning. Finally, the predictive performance of these models will be rigorously evaluated using standard regression metrics, primarily the Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), and R-squared score, to identify the most reliable forecasting algorithm for volatile market conditions.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
try:
    # 1. Create a deep copy of the original dataset to avoid modifying the baseline source data
    df_cleaned = dataset.copy()
    
    # 2. Parse the string date column into a robust standard datetime format ('MMM-YY' format like 'Jul-05')
    df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], format='%b-%y')
    
    # 3. Sort the dataset chronologically by Date to maintain correct temporal sequence for time-series modeling
    df_cleaned = df_cleaned.sort_values(by='Date').reset_index(drop=True)
    
    print("✅ Data Wrangling Complete: Date column converted and rows sorted chronologically.")
    print("\nVerification of Head:")
    display(df_cleaned.head())
    
    # 4. Extract seasonal components for later EDA use (useful feature engineering)
    df_cleaned['Year'] = df_cleaned['Date'].dt.year
    df_cleaned['Month'] = df_cleaned['Date'].dt.month
    print("✅ Feature extraction complete: 'Year' and 'Month' columns appended for visualization.")
    
except NameError:
    print("❌ Error: 'dataset' is not defined. Please run the data loading cell first.")
except Exception as e:
    print(f"❌ An error occurred during the Data Wrangling process: {e}")
````

### What all manipulations have you done and insights you found?

During the data wrangling phase, the following transformations and structural manipulations were executed systematically:

1. Source Data Preservation: A deep copy of the baseline dataframe was created to preserve the original raw records and avoid any in-place side-effect conflicts.

2. Temporal Parsing: The Date column was successfully transformed from a generic text object into a standardized, rich datetime64[ns] format. This explicitly allows Python to recognize month-over-month intervals.

3. Chronological Realignment: The rows were sorted in ascending order based on the parsed dates and the index was reset. This ensures the data acts as a cohesive time-series sequence where preceding rows directly feed into succeeding observations.

4. Feature Extraction: Two new features, Year and Month, were explicitly extracted from the synchronized timestamp index.

4. Segmented Risk Management: The time-series trendline clearly maps out a severe structural break (the 2018 liquidity crisis).
